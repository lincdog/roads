```python
import geopandas as gpd
import pandas as pd
#import numpy as np
#import matplotlib.pyplot as plt
from ipywidgets import interact, fixed, SelectMultiple
import re
from IPython.display import display

pd.set_option('display.max_columns', None)
```

```python
df = gpd.read_file('/Users/lincoln/geography/massachusetts/MassDOT_Roads_SHP/ma_num_routes_union.shp')
```

```python
def rt_sort_key(x):
    return int(re.findall('\d+', x)[0])

RT_NUMBERS = sorted(df['RT_NUMBER'].unique(), key=rt_sort_key)
```

```python
interstates = ['90', '190', '290', '95', '495', '195', '295', '91', '93']
```

```python
df_nointers = df[~df['RT_NUMBER'].isin(interstates)]
```

```python
df_nointers.sort_values(by='length', ascending=False).query('length > 50')
```

```python
def display_route(rt, df):
    colors = ['blue', 'red', 'green', 'orange', 'purple', 'black', 'yellow']
    n_colors = len(colors)
    
    rt_entry = df.loc[df['RT_NUMBER']==rt[0]]
    length = rt_entry['length'].values[0]
    m = rt_entry.explore(tiles='CartoDB positron', width=1000, height=400)
    print(f'{rt[0]} : {length} miles')
    
    for i in range(1, len(rt)):
        rt_entry = df.loc[df['RT_NUMBER']==rt[i]]
        length = rt_entry['length'].values[0]
        rt_entry.explore(m=m, color=colors[i % n_colors])
        print(f'{rt[i]} : {length} miles')

    return m
```

```python
interact(
    display_route, 
    rt=SelectMultiple(description='Route(s)', options=list(RT_NUMBERS), value=["1"]),
    df=fixed(df)
)
```
